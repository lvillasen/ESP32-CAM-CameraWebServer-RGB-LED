```python
from PIL import Image
import matplotlib.pyplot as plt
import requests
import time
from io import BytesIO
url='http://192.168.1.86'
fig = plt.figure(figsize=(10, 7)) 
time.sleep(2)
# Image 1 with white LED
response = requests.get(url+ "/ledRGB?ledRed=255&ledGreen=255&ledBlue=255")
time.sleep(1)
response = requests.get(url+ "/capture")
img = Image.open(BytesIO(response.content))
fig.add_subplot(2, 2, 1) 
plt.imshow(img) 
plt.title("White") 
time.sleep(.1)

# Image 2 with red LED
response = requests.get(url+ "/ledRGB?ledRed=255&ledGreen=0&ledBlue=0")
time.sleep(1)
response = requests.get(url+ "/capture")
img = Image.open(BytesIO(response.content))
fig.add_subplot(2, 2, 2) 
plt.imshow(img)
plt.title("Red") 
time.sleep(.1)

# Image 3 with green LED
response = requests.get(url+ "/ledRGB?ledRed=0&ledGreen=255&ledBlue=0")
time.sleep(1)
response = requests.get(url+ "/capture")
img = Image.open(BytesIO(response.content))
fig.add_subplot(2, 2, 3) 
plt.imshow(img)
plt.title("Green") 
time.sleep(.1)

# Image 4 with blue LED
response = requests.get(url+ "/ledRGB?ledRed=0&ledGreen=0&ledBlue=255")
time.sleep(1)
response = requests.get(url+ "/capture")
img = Image.open(BytesIO(response.content))
fig.add_subplot(2, 2, 4) 
plt.imshow(img)
plt.title("Blue") 
plt.show()
```
